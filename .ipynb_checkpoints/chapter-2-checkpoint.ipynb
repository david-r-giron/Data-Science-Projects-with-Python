## 2. Introduction to Scikit-Learn

```python
from sklearn.linear_model import LogisticRegression
```

#### Logistic regression, despite its name, is actually a classification model. This is one of the simplest (and therefore most important) classification models.

```python
my_lr = LogisticRegression()
my_lr
```

#### Any time you use a machine learning model that has been prepared for you as scikit-learn models have been, your first job is to understand each and every option that is available.

#### A best practice in such cases is to explicitly provide every keyword parameter to the model when you create the object, even if you are just selecting all the default options. This will help increase your awareness of the decisions that are being made.

```python
my_new_lr = LogisticRegression(
    C=1.0, 
    class_weight=None, 
    dual=False, 
    fit_intercept=True,
    intercept_scaling=1, 
    max_iter=100, 
    multi_class='auto',
    n_jobs=None, 
    penalty='l2', 
    random_state=None, 
    solver='warn',
    tol=0.0001, 
    verbose=0, 
    warm_start=False
)
my_new_lr
```

#### You can also just instantiate with the default options, as we just did above, or simply without specifying them, and then make changes later as necessary.

```python
my_new_lr.C = 0.1 # change from the original default 1.0
my_new_lr.solver = 'liblinear' # originally 'warn'
my_new_lr
```

The <strong>-1</strong> in the first positional argument of <code>.reshape</code> means to make the output array shape flexible in that dimension, according to how much data goes in. Rows will be filled while at first we only have one column.</p>

```python
y = df['default payment next month'][0:10].values
y
```

## 3. Generating Synthetic Data

<p>We use NumPy to make a one-dimensional array of feature values X, consisting of 1,000 random real numbers between 0 and 10. Next, we use <code>numpy.random.uniform</code> which draws from the uniform distribution: it's equally likely to choose any number between <strong>low</strong> (inclusive) and <strong>high</strong> (exclusive), and will return an array of whatever <strong>size</strong> you specify.</p>

```python
np.random.seed(seed=1)
X = np.random.uniform(low=0.0, high=10.0, size=(1000,))
X[0:10]
```

```python
type(X)
```

## 4. Data for a Linear Regression

<p>Now we need a response variable. We'll generate data that follows <strong>the assumptions of linear regression: data that follows a linear trend but has normally distributed errors.</strong> Note that it is unlikely that a real-world data set would satisfy the formal statistical assumptions of a model like this.</p>

<p>Linear Regression theoretically should only be used to model data where the reponse is a linear transformation of the features, with normally distributed (Gaussian) noise:</p>

<p>y = ax + b + N(μ, σ)</p>

<p><em>a</em> is the slope, <em>b</em> is the intercent, and the Gaussian noise has a mean of <em>μ</em> with a standard deviation of <em>σ</em>.</p>

```python
np.random.seed(seed=1)
slope = 0.25 # or 'a'
intercept = -1.25 # or 'b'

# loc is the mean, scale is the standard deviation, and size is the number of data points
gaussian_noise = np.random.normal(loc=0.0, scale=1.0, size=(1000,))

y = slope * X + intercept + gaussian_noise 
```

```python
mpl.rcParams['figure.dpi'] = 400 # dpi = dots per inch
plt.scatter(X,y,s=1)
```

## 5. (Exercise 8): Linear Regression in Scikit-Learn

<p>Notice that for a linear regression, there are not that many options to specify... The default settings include <strong>fit_intercept=True</strong>, meaning the regression model will include an intercept term.</p>

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg
```

<p>Now we can fit the model using our synthetic data, remembering to reshape the feature array so that the samples are along the first dimension. </p>

```python
lin_reg.fit(X.reshape(-1,1), y)
```

<p>After fitting the linear regression model, we examine <code>lin_reg.intercept_</code> and <code>lin_reg.coef_</code>, which contain intercept and slope coefficient.</p>

```python
print(lin_reg.intercept_)
print(lin_reg.coef_)
```

<p>Finally, we can use the model to make predictions on feature values. Here, we do this using the same data used to fit the model: the array of features, <strong>X</strong>. We capture the output of this as a variable, <strong>y_pred</strong>.</p>

```python
y_pred = lin_reg.predict(X.reshape(-1,1))
```

#### Scatter plot of X and y, plus a line plot of the predicted values. For the line plot we add 'r' to have a red colored line. For the scatter plot, the parameter s is the size of markers, a multiplier of their default area.

```python
plt.scatter(X,y,s=1)
plt.plot(X,y_pred,'r')
```

## 6. Model performance metrics for binary classification

### Splitting the data: training and testing sets

#### Split the data in to training (80%) and testing (20%). Introduce metrics using testing scores.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df['EDUCATION'].values.reshape(-1,1), # X
    df['default payment next month'].values, # y
    test_size=0.2, 
    random_state=24
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
np.mean(y_train)
```

```python
np.mean(y_test)
```

## 7. Classification accuracy

```python
from sklearn.linear_model import LogisticRegression

example_lr = LogisticRegression(
    C=0.1, 
    class_weight=None, 
    dual=False, 
    fit_intercept=True,
    intercept_scaling=1, 
    max_iter=100, 
    multi_class='auto',
    n_jobs=None, 
    penalty='l2', 
    random_state=None, 
    solver='liblinear',
    tol=0.0001, 
    verbose=0, 
    warm_start=False
)

example_lr.fit(X_train, y_train)
y_pred = example_lr.predict(X_test)
```

#### Now to measure accuracy in a binary classification setting

```python
is_correct = (y_pred == y_test)
np.mean(is_correct)
```

<strong>Now using the <code>score</code> method.</strong>

```python
example_lr.score(X_test, y_test)
```

<strong>And now using the <code>accuracy_score</code> methd from the <code>metrics</code> package</strong>

```python
from sklearn import metrics

metrics.accuracy_score(y_test, y_pred)
```

## 8. (Exercise 9): Calculating the True and False Positive and Negative Rates and Confusion Matrix in Python

#### Positives

```python
P = sum(y_test)
P
```

#### True Positives

```python
TP = sum((y_test == 1) & (y_pred==1))
TP
```

#### True Positives Rate

```python
TPR = TP / P
TPR
```

#### False Negatives

```python
FN = sum( (y_test == 1) & (y_pred == 0) )
FN
```

#### False Negatives Rate

```python
FNR = FN/P
FNR
```

#### Negatives

```python
N = sum(y_test==0)
N
```

#### True Negatives

```python
TN = sum( (y_test==0) & (y_pred==0))
TN
```

#### True Negatives Rate

```python
TNR = TN/N
TNR
```

#### False Positives

```python
FP = sum( (y_test==0) & (y_pred==1))
FP
```

#### False Positives Rate

```python
FPR = FP/N
FPR
```

### Confusion Matrix

<p>[ True Negative,  False Positive</p>
<p> False Negatives,  True Positive]</p>

```python
from sklearn import metrics

metrics.confusion_matrix(y_test, y_pred)
```

## 9. (Exercise 10): Obtainign Predicted Probabilities from a Trained Logistic Regressions Model

<p> One thing to understand about how logistic regression works is that the raw predictions (the direct outputs from the mathematical equation that defines logistic regression) are not binary labels. They are actually probabilities on a scale from 0 to 1, exclusive on both ends. These probabilities are only transformed into binary predictions through the use of a <strong>threshold</strong>.</p>

```python
y_pred_proba = example_lr.predict_proba(X_test)
y_pred_proba
```

<p>Above we see two columns. This is because there are two classes in our classification problem: negative and positive. Assuming the negative labels are coded as 0 and the positive as 1, as they are in our data, scikit-learn will report the probabilitiy of negative class membership as the first column, and positive class membership as the second.</p>

### Visualize the predicted probability of positive class

```python
pos_proba = y_pred_proba[:,1]

mpl.rcParams['figure.dpi'] = 80
mpl.rcParams['font.size'] = 9

plt.hist(pos_proba)
plt.xlabel('Predicted probability of positive class for testing data')
plt.ylabel('Number of samples')
```

<p>Notice how there are only four bins that actually have samples in them, and they are spaced fairly apart. This is because there are only four unique values for the <strong>EDUCATION</strong> feature, which is the only feature in our example model.</p>

<p>Also notice that all the predicted probabilities are below 0.5. This is the reason every sample was predicted to be negative, using the 0.5 threshold. If we set our threshold at a lower number, like 0.25, then we would get different results, though not necessarily more accurate.</p>

### Negative and Positive samples for predicted porbability for positive class

```python
pos_sample_pos_proba = pos_proba[y_test==1]
neg_sample_pos_proba = pos_proba[y_test==0]

mpl.rcParams['figure.dpi'] = 80
mpl.rcParams['font.size'] = 9

plt.hist([pos_sample_pos_proba, neg_sample_pos_proba], histtype='barstacked')
plt.legend(['Positive samples', 'Negative samples'])
plt.xlabel('Predicted probability of positive class')
plt.ylabel('Number of samples')
```

## 10. The Receiver Operating Charasteristic (ROC) Curve

<p>The receiver operating characteristic (ROC) curve is a plot of the pairs of true positive rates (y-axis) and false positive rates (x-axis) that result from lowering the threshold down from 1, all the way to 0.</p>

```python
from sklearn import metrics

fpr, tpr, thresholds = metrics.roc_curve(y_test, pos_proba)

mpl.rcParams['figure.dpi'] = 100
mpl.rcParams['font.size'] = 12

plt.plot(fpr, tpr, '*-')
plt.plot([0, 1], [0, 1], 'r--')
plt.legend(['Logistic regression', 'Random chance'])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve')
```

```python
thresholds
```

```python
metrics.roc_auc_score(y_test, pos_proba)
```

## 11. (Activity 2): Performing Logistic Regression with a New Feature and Creating a Precision-Recall Curve

```python
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(
    df['LIMIT_BAL'].values.reshape(-1,1), 
    df['default payment next month'].values,
    test_size=0.2, 
    random_state=24
)

example_lr.fit(X_train_2, y_train_2)

y_test_2_pred_proba = example_lr.predict_proba(X_test_2)

metrics.roc_auc_score(y_test_2, y_test_2_pred_proba[:,1])
```

```python
from sklearn import metrics

fpr_2, tpr_2, thresholds_2 = metrics.roc_curve(y_test_2, y_test_2_pred_proba[:,1])

mpl.rcParams['figure.dpi'] = 100
mpl.rcParams['font.size'] = 12

plt.plot(fpr_2, tpr_2, '*-')
plt.plot([0, 1], [0, 1], 'r--')
plt.legend(['Logistic regression', 'Random chance'])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve for logistic regression with LIMIT_BAL feature')
```

<p>The ROC curve advances as the threshold decreases(?)</p>

```python
thresholds_2
```

### Precision and Recall

```python
precision, recall, thresh_3 = metrics.precision_recall_curve(y_test_2, y_test_2_pred_proba[:,1])

plt.plot(recall, precision, '-x')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision and recall for the logistic regression with LIMIT_BAL')
plt.xlim([0, 1])
plt.ylim([0, 1])
```
